```
import pandas as pd
import random

# Leitura do arquivo:
arquivo = pd.read_csv('heart.csv', sep = ",");
#arquivo = arquivo.drop([0])
#arquivo

pontos = arquivo.drop(columns=["sex", "cp", "trestbps", "fbs", "restecg", "exang", "oldpeak", "slope", "ca", "thal"], axis=1);

lista = pontos.values
random.shuffle(lista)
#lista

#pontos = pd.DataFrame(lista);
#pontos

pontos

#print(pontos.iloc[2]["target"]);

#for i in range (len(pontos)):
#  if(pontos.iloc[i]["target"] == 0):
#    print(str(i) + ' é 0')


#for i in range (len(pontos)):
#  if(pontos.iloc[i]["target"] == 1):
#    print(str(i) + ' é 1')
```

```
from sklearn.model_selection import train_test_split

treino, teste = train_test_split(pontos, test_size=0.30);

treino

#teste
```

```
import math
import statistics

# 1. Definir um valor fixo para K
k = 5;


#
def distancia(cluster_antigo, cluster_novo):
  x = float(cluster_antigo[0]) - cluster_novo[0];
  x = x * x;
  y = float(cluster_antigo[1]) - cluster_novo[1];
  y = y * y;

  d = math.sqrt(x + y)
  return d;
#

def predict_KNN(pontosTreinamento, ponto):
  
  tabela = [];

  #2. Calcular a distância entre o ponto que desejamos classificar e todos os pontos do dataset.
  dist = 0;
  for j in range (len(pontosTreinamento)):
    dist = distancia( [float(pontosTreinamento.iloc[j][0]), float(pontosTreinamento.iloc[j][1]) ], [ponto[0], ponto[1]]);
    #print(dist);

    p = pontosTreinamento.iloc[j];
    p = p[3];
    #print(p);

    #3. Armazenar as distâncias calculadas em uma tabela.
    tabela.append([dist, p]);
    

  #4. Ordenar a tabela em ordem crescente de acordo com a distância calculada.
  tabela = sorted(tabela);
  #print(tabela);


  #5. Selecionar as labels das K primeiras entradas da tabela.
  df = pd.DataFrame(tabela[0:k]);
  #print(df);
  
  # 6. A label que for a moda dessas K primeiras entradas será a classe predita para o ponto em análise.
  labels = df[1].tolist();
  
  #print(labels);
  moda = statistics.mode(labels);


  print('\n\n');

  #return 1;
  return moda;

  
```

```

print('\nINÍCIO DO TREINAMENTO\n\n');

tabelaRetornoTreinamento = [];

for i in range (len(treino)):
#for i in range (3):

  retornoTreinamento = predict_KNN(treino, [treino.iloc[i][0], treino.iloc[i][1]] );

  tabelaRetornoTreinamento.append([treino.iloc[i][0], treino.iloc[i][1], retornoTreinamento]);
  
  print('Indice: ' +str(i) + ' é da target = ' + str(treino.iloc[i][3]))
  print('A classe retornada é: ' + str(retornoTreinamento) + '\n');
  print('\n');



print('\nFIM DO TREINAMENTO!\n\n\n');
```

```
import math
import statistics

# 1. Definir um valor fixo para K
k = 5;


#
def distancia(cluster_antigo, cluster_novo):
  x = float(cluster_antigo[0]) - cluster_novo[0];
  x = x * x;
  y = float(cluster_antigo[1]) - cluster_novo[1];
  y = y * y;

  d = math.sqrt(x + y)
  return d;
#

def predict_KNN_Regressao(pontosTreinamento, ponto):
  
  tabela = [];

  #2. Calcular a distância entre o ponto que desejamos classificar e todos os pontos do dataset.
  dist = 0;
  for j in range (len(pontosTreinamento)):
    dist = distancia( [float(pontosTreinamento.iloc[j][0]), float(pontosTreinamento.iloc[j][1]) ], [ponto[0], ponto[1]]);
    #print(dist);

    p = pontosTreinamento.iloc[j];
    p = p[2];
    #print(p);

    #3. Armazenar as distâncias calculadas em uma tabela.
    tabela.append([dist, p]);
    

  #4. Ordenar a tabela em ordem crescente de acordo com a distância calculada.
  tabela = sorted(tabela);
  #print(tabela);


  #5. Selecionar as labels das K primeiras entradas da tabela.
  df = pd.DataFrame(tabela[0:k]);
  #print(df);
  
  # 6. A label que for a moda dessas K primeiras entradas será a classe predita para o ponto em análise.
  valores = sum(df[1].tolist()) / k;
  
  print('\n\n');

  return valores;
```

```

print('\nINÍCIO DO TREINAMENTO\n\n');

tabelaRetornoTreinamento = [];

for i in range (len(treino)):

  retornoTreinamento = predict_KNN_Regressao(treino, [treino.iloc[i][0], treino.iloc[i][1]] );

  print('Valor retornado: ' + str(retornoTreinamento) );
  print('Valor correto: ' + str(treino.iloc[i][2]));
  print('\n');

  tabelaRetornoTreinamento.append([treino.iloc[i][0], treino.iloc[i][1], treino.iloc[i][2], retornoTreinamento]);

print('\nFIM DO TREINAMENTO!\n\n\n');
```
